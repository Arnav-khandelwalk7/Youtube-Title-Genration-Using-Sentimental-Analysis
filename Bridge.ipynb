```python
import os
import operator
import pandas as pd
```

```python
data = pd.read_csv('comments.csv')
data
```

```python
sorted_data = data.sort_values(by="comment_like_count", ascending=False)
```

```python
top_sorted = sorted_data.head(500)
top_sorted
```

```python
polarity_sorted = top_sorted[top_sorted['polarity_category'] == 1]
polarity_sorted
```

```python
title = polarity_sorted['video_title'].value_counts().keys().tolist()
frequency = polarity_sorted['video_title'].value_counts().tolist()

fusion = list(zip(title, frequency))

title_frame = pd.DataFrame(fusion, columns=['title', 'frequency']).head(10)
title_frame
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vect = CountVectorizer()
tf_train = vect.fit_transform(title_frame['title'].values.astype('U'))
tf_test = vect.transform(title_frame['frequency'].values.astype('U'))
```

```python
vector_vocab = vect.vocabulary_
vector_vocab
```

```python
sorted_vocab = dict(sorted(vector_vocab.items(), key=operator.itemgetter(1), reverse=True))
sorted_vocab
```

```python
import spacy
nlp = spacy.load("en_core_web_sm")
cnt = 0
final = []
for  i in vector_vocab.keys():
  x = nlp(i)
  if(x[0].tag_  == 'NNP'):
    final.append(i);
    print(i, " is a noun.")
    cnt = cnt + 1
    if(cnt == 6):
      break
  
```
