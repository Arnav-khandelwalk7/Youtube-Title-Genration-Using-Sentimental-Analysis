```python
import re
import nltk
import demoji
import pandas as pd
from langdetect import detect
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from youtubesearchpython import VideosSearch
```

```python
import csv
import pandas
colnames = ['num','val']
data = pandas.read_csv('keywords.csv', names=colnames)
data
```

```python
words_list = data.val.tolist()
```

```python
words_list.remove('0')
words_list
```

```python
title_list = []
limit = 0

for i in words_list:
  search = VideosSearch(i)
  limit += 100
  while len(title_list) <= limit:
      stop = False
      search_result = search.result()['result']    
      for x in search_result:
          if len(title_list) != limit:
              title_list.append(x['title'])
          else:
              stop = True  
      if stop:
          break
      else:
          try:
              search.next()
          except:
              stop = True
              break
for index, title in enumerate(title_list):
    print(index, title)
```

```python
data = pd.DataFrame(title_list, columns=['video_title'])
data
```

```python
# Removing Emojis
data['video_title'] = data['video_title'].apply(lambda x: demoji.replace(x, ''))
data
```

```python
# Language Detection
def try_detect(title):
    language = ''
    try:
        language = detect(title)
    except:
        language = 'error'
    return language

data['language'] = data['video_title'].apply(lambda x: try_detect(x))
data
```

```python
# Extracting the english comments only
data = data[data['language'] == 'en']
data = data.drop(['language'], axis=1)
data
```

```python
# Removing the Special Characters
regex = r'[^0-9A-Za-z\t]'
video_title = data.loc[:, 'video_title']
video_title = video_title.apply(lambda x: re.sub(regex, ' ', x))
data = data.drop(['video_title'], axis=1)
data['video_title'] = video_title
data
```

```python
# Lower case conversion
data['video_title'] = data['video_title'].str.lower()
data
```

```python
# Stripping trailing spaces
data['video_title'] = data['video_title'].str.strip()
data
```

```python
# Removing stopwords
nltk.download('punkt')
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

def remove_stopwords(line):
    word_tokens = word_tokenize(line)
    filtered_sentence = [w for w in word_tokens if not w in stop_words]
    return ' '.join(filtered_sentence)
```

```python
data['stop_title'] = data['video_title'].apply(lambda x: remove_stopwords(x))
data
```

```python
data.to_csv('search_results.csv', index=False)
```
